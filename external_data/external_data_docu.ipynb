1. **`numer_sta`**: 
   - Station number or ID where the measurements were recorded.

2. **`date`**: 
   - Date and time of the observation in `YYYY-MM-DD HH:MM:SS` format.

3. **`pmer`**: 
   - Sea-level pressure (in Pascals or hPa, depending on units).

4. **`tend`**: 
   - Pressure tendency (change in pressure over a certain period, e.g., last 3 hours).

5. **`cod_tend`**: 
   - Code indicating the type of pressure tendency (e.g., rising, falling).

6. **`dd`**: 
   - Wind direction (in degrees).

7. **`ff`**: 
   - Wind speed (likely in m/s or km/h).

8. **`t`**: 
   - Temperature (likely in Kelvin or Celsius).

9. **`td`**: 
   - Dew point temperature (indicating moisture content in the air).

10. **`u`**: 
    - Relative humidity (percentage).

11. **`vv`**: 
    - Visibility distance (likely in meters or kilometers).

12. **`ww`, `w1`, `w2`**: 
    - Weather codes (e.g., precipitation types, storms, or other weather phenomena).

13. **`n`**: 
    - Total cloud cover (in octas, where 0 = clear and 8 = overcast).

14. **`nbas`**: 
    - Fraction of cloud cover at the lowest level.

15. **`hbas`**: 
    - Height of the cloud base (likely in meters or feet).

16. **`cl`**: 
    - Type of low clouds.

17. **`cm`**: 
    - Type of medium clouds.

18. **`ch`**: 
    - Type of high clouds.

19. **`pres`**: 
    - Atmospheric pressure at the station.

20. **`niv_bar`**: 
    - Barometric level (possibly indicating altitude or pressure level).

21. **`geop`**: 
    - Geopotential height (altitude above a reference level).

22. **`tend24`**: 
    - Temperature tendency over the past 24 hours.

23. **`tn12`, `tn24`**: 
    - Minimum temperature over the last 12 and 24 hours, respectively.

24. **`tx12`, `tx24`**: 
    - Maximum temperature over the last 12 and 24 hours, respectively.

25. **`tminsol`**: 
    - Minimum soil temperature.

26. **`sw`**: 
    - Surface wetness or solar radiation.

27. **`tw`**: 
    - Wet bulb temperature.

28. **`raf10`**: 
    - Gust speed over the last 10 minutes.

29. **`rafper`**: 
    - Gust speed over a specified period.

30. **`etat_sol`**: 
    - State of the ground (e.g., wet, frozen, snow-covered).

31. **`ht_neige`**: 
    - Snow height.

32. **`ssfrai`**: 
    - Snowfall during a recent period.

33. **`rr1`, `rr3`, `rr6`, `rr12`, `rr24`**: 
    - Precipitation amounts over the last 1, 3, 6, 12, and 24 hours, respectively.

34. **`phenspe1` to `phenspe4`**: 
    - Special weather phenomena observed (e.g., fog, storms).

35. **`nnuage1` to `nnuage4`**: 
    - Cloud cover at different levels.

36. **`ctype1` to `ctype4`**: 
    - Cloud types at different levels.

37. **`hnuage1` to `hnuage4`**: 
    - Height of clouds at different levels.
